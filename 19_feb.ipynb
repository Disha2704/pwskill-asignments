Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible Python web framework that provides a simple and elegant way to build web applications. It was developed by Armin Ronacher in 2010 and is based on the Werkzeug WSGI toolkit and the Jinja2 template engine.

Some advantages of Flask framework are:

Lightweight: Flask is a lightweight framework, which means it doesn't require particular tools or libraries. This makes it easier to learn and use.

Flexibility: Flask provides a lot of flexibility to developers, allowing them to create web applications in any way they want.

Modular design: Flask has a modular design, which means that it has a core framework and extensions that can be used to add more functionality.

Easy to get started: Flask has a simple and intuitive API that makes it easy to get started with web development.

Extensible: Flask is highly extensible, meaning that developers can add new features and functionality to their applications easily.

Built-in development server: Flask comes with a built-in development server that allows developers to test their applications locally.

Large community: Flask has a large and active community of developers who contribute to its development, provide support, and create extensions that can be used in applications.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, routing refers to the process of binding a URL to a function that handles the request for that URL. It defines how the application responds to requests from the client based on the URL requested by the client. The process of routing in Flask is implemented using the @app.route() decorator.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL for the specified endpoint with the given arguments. The url_for() function takes the name of the endpoint as its first argument and any number of keyword arguments representing the values to be substituted in the URL.
